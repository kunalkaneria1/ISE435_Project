```python
# function for User Input on initial launch
def UserIn():
    sg.theme('BlueMono')   # Add a touch of color
    # All the stuff inside your window.
    layout = [  [sg.Text('Welcome To PC Building!')],
                [sg.Text('Would you like to build?: Y/N'), sg.InputText()],
                [sg.Button('Ok')] ]

    # Create the Window
    window = sg.Window('Welcome To PC Building!', layout)
    # Event Loop to process "events" and get the "values" of the inputs
    while True:
        event, values = window.read()
        if event == sg.WIN_CLOSED or event == 'Ok': # if user closes window/clicks cancel/goes ahead
            break
        

    window.close()
    
    return values #return input
```

```python
# Main GUI Code Block
def GUI(CPU, MOB, GPU, PSU, MEM, HDD, SSD, COL, budget):
    
    currDF = CPU #Set the working table DF to CPU as default
    filtDF = currDF #Set filtDF to currDF for filter system
    
    outDF = pd.DataFrame() #Need to create output DF for parts to be put in
    outDF['Item Name'] = ['Enter Parts Below!'] #Code doesn't like it if columns aren't made for DF in table
    outDF['MSRP Price'] = [0]
    
    sg.theme('BlueMono')   # Add a touch of color
    # All the stuff inside your window.
    layout = [  [sg.Text('Filter your results or pick a parts list!')],
                [sg.Text('Enter A Minimum Price:'), sg.InputText(key='min')], #Set minimum for filter
                [sg.Text('Enter A Maximum Price:'), sg.InputText(key='max')], #Set max for filter
                [sg.Text('Enter A Keyword:'), sg.InputText(key='search')], #Set a search word
                [sg.Button('Filter')],
                [sg.Text('', key='errorcode1')], #Error code for filter system
                [sg.Text('Pick any button to change lists')], #Create buttons for table swapping
                [sg.Button('CPU'),
                 sg.Button('MotherBoard'),
                 sg.Button('GPU'),
                 sg.Button('PSU'),
                 sg.Button('MEM'),
                 sg.Button('HDD'),
                 sg.Button('SSD'),
                 sg.Button('Cooler')],
                [sg.Table(values=currDF.values.tolist(), #Table for displaying parts DFs
                          headings=['Part', 'Price'],
                          display_row_numbers=True,
                          auto_size_columns=False,
                          col_widths=[90, 7],
                          num_rows=8,
                          key='_TABLE_')],
                [sg.Text('Buy or Remove A Part? Enter Row Index:'), sg.InputText(key='index')], #Text/Input for Indexing
                [sg.Button('Finalize List'), sg.Button('Add Part'), sg.Button('Remove Part')], #Buttons for filter/output DF interacting
                [sg.Text('', key='errorcode2')], #Error code for incorrect index 
                [sg.Text('Parts Selected:')],
                [sg.Table(values=outDF.values.tolist(), #table for displaying output DF list
                          headings=['Part', 'Price'],
                          display_row_numbers=True,
                          auto_size_columns=False,
                          col_widths=[90, 7],
                          num_rows=8,
                          key='_TABLE2_')],
                [sg.Text('Current Amount:'), sg.Text(outDF['MSRP Price'].sum(), key='sum')], #display current total in output DF
                [sg.Text('Budget Set:'), sg.Text(budget)] #display budget
                ]

    # Create the Window
    window = sg.Window('Welcome To PC Building!', layout=layout, resizable=True, size=(960, 780), finalize=True, 
                       auto_size_text=True, auto_size_buttons=True) #set size for window
    window['_TABLE_'].expand(True, True) #set expansion of both tables as true
    window['_TABLE2_'].expand(True, True)
    
    # Event Loop to process "events" and get the "values" of the inputs
    while True:
        event, values = window.read()
        if event == sg.WIN_CLOSED: #Break if user closes window
            break
        
        elif event == 'Finalize List': #Break to return DF back out
            window.close()
            return outDF
        
        # If statement chain for all possible button presses
        
        elif event == 'Filter': # if user chooses to filter the DF
            if (values['min'] == '') or (values['max'] == ''): #Check if max or min is empty
                
                filtDF = pd.DataFrame() #reset filtDF
                filtDF = filtDF.append(currDF.loc[currDF['Item Name'].str.contains(values['search'])], ignore_index=True)
                filtDF = filtDF.sort_values('MSRP Price', ignore_index=True) #Set filter to just search keyword and sort
                window['_TABLE_'].update(values=filtDF.values.tolist()) #update table
                window['errorcode1'].update('') #update error code to none
                
            else:
                try: #test to make sure if min and max are real numbers
                    int(values['min'])
                    int(values['max'])
                    
                    filtDF = pd.DataFrame() #reset filtDF
                    
                    #use search keyword, min, and max to create filtDF
                    filtDF = filtDF.append(currDF.loc[(currDF['Item Name'].str.contains(values['search'])) &
                                                      (currDF['MSRP Price'] >= int(values['min'])) & 
                                                      (currDF['MSRP Price'] <= int(values['max']))], ignore_index=True)
                    filtDF = filtDF.sort_values('MSRP Price', ignore_index=True) #sort DF 
                    window['_TABLE_'].update(values=filtDF.values.tolist()) #update table
                    window['errorcode1'].update('') #update error code to none
                    
                except ValueError: #if error
                    window['errorcode1'].update('Please enter in correct integer values.') #error code for incorrect values
            
        elif event == 'Add Part': # if user decides to add a part to the outDF
            try: #test to make sure index is a legitimate number to prevent crash
                int(values['index'])
                
                outDF = outDF.append(filtDF.loc[int(values['index'])], ignore_index=True) #add index from filtDF to outDF, ignore index
                window['_TABLE2_'].update(values=outDF.values.tolist()) #update output table
                window['errorcode2'].update('') #update error code to none
                window['sum'].update(outDF['MSRP Price'].sum()) #update running sum
                
            except ValueError: #if error on proper int
                window['errorcode2'].update('Please enter in correct integer values.') #update error code for incorrect values
        
        elif event == 'Remove Part': # if user decides to remove a part from outDF
            try: #test to make sure index is a legitimate number to prevent crash
                int(values['index'])
                
                outDF = outDF.drop(int(values['index'])) #removes row from outDF
                outDF = outDF.reset_index(drop=True) #resets index to be sequential
                window['_TABLE2_'].update(values=outDF.values.tolist()) #update out table
                window['errorcode2'].update('') #update error code to none
                window['sum'].update(outDF['MSRP Price'].sum()) #update running sum
                
            except ValueError:
                window['errorcode2'].update('Please enter in correct integer values.')
        
        elif event == 'CPU': # Switch DF to CPU DF
            currDF = CPU
            filtDF = currDF
            window['_TABLE_'].update(values=currDF.values.tolist())
            
        elif event == 'MotherBoard': # Switch DF to Motherboard DF
            currDF = MOB
            filtDF = currDF
            window['_TABLE_'].update(values=currDF.values.tolist())
            
        elif event == 'GPU': # Switch DF to GPU DF
            currDF = GPU
            filtDF = currDF
            window['_TABLE_'].update(values=currDF.values.tolist())
            
        elif event == 'PSU': # Switch DF to PSU DF
            currDF = PSU
            filtDF = currDF
            window['_TABLE_'].update(values=currDF.values.tolist())
            
        elif event == 'MEM': # Switch DF to MEM DF
            currDF = MEM
            filtDF = currDF
            window['_TABLE_'].update(values=currDF.values.tolist())
            
        elif event == 'HDD': # Switch DF to HDD DF
            currDF = HDD
            filtDF = currDF
            window['_TABLE_'].update(values=currDF.values.tolist())
            
        elif event == 'SSD': # Switch DF to SSD DF
            currDF = SSD
            filtDF = currDF
            window['_TABLE_'].update(values=currDF.values.tolist())
            
        elif event == 'Cooler': # Switch DF to Cooler DF
            currDF = COL
            filtDF = currDF
            window['_TABLE_'].update(values=currDF.values.tolist())

    window.close()
    
    
    
```

```python
# general purpose function for choosing cooling type and platform
def PickAPart(text):
    sg.theme('BlueMono')   # Add a touch of color
    # All the stuff inside your window.
    layout = [  [sg.Text(text), sg.InputText()], #input for checking platform or cooling solution
                [sg.Button('Ok')] ]

    # Create the Window
    window = sg.Window('Welcome To PC Building!', layout)
    # Event Loop to process "events" and get the "values" of the inputs
    while True:
        event, values = window.read()
        if event == sg.WIN_CLOSED or event == 'Ok': # if user closes window/clicks cancel/goes ahead
            break
        

    window.close()
    
    return values
```

```python
# Leave screen once code ends
def LeaveScreen():
    sg.theme('BlueMono')   # Add a touch of color
    # All the stuff inside your window.
    layout = [  [sg.Text('Thank You!')],
                [sg.Button('Ok')] ]

    # Create the Window
    window = sg.Window('Welcome To PC Building!', layout)
    # Event Loop to process "events" and get the "values" of the inputs
    while True:
        event, values = window.read()
        if event == sg.WIN_CLOSED or event == 'Ok': # if user closes window/clicks cancel/goes ahead
            break
        

    window.close()
    
    
```

```python
#general purpose function for input errors
def ErrorCall():
    sg.theme('BlueMono')   # Add a touch of color
    # All the stuff inside your window.
    layout = [  [sg.Text('Something went wrong! Please input again')],
                [sg.Button('Ok')] ]

    # Create the Window
    window = sg.Window('Uh Oh!', layout)
    # Event Loop to process "events" and get the "values" of the inputs
    while True:
        event, values = window.read()
        if event == sg.WIN_CLOSED or event == 'Ok': # if user closes window/clicks cancel/goes ahead
            break

    window.close()
    
    
```

```python
#Function for creating DataFrames of parts
#Passes the link into function, returns the DF to clone variable
def ReadLink(weblink):
    
    #Arrays for names, prices, and item comparison
    name_list = []
    price_list = []
    item_list = []
    
    driver.get(weblink) # push driver to link first page
    
    pageNum = driver.find_element(By.XPATH, '//span[@class="list-tool-pagination-text"]//strong').text #fetch page number for iteration
    pageNum = pageNum.replace('1/', "") #gets rid of the 1/ in 1/#
    
    for i in range(1, int(pageNum)+1): #for loop to iterate through pages
        driver.get(weblink+'page='+str(i)) # push driver to pages based on how many there are, iterate through each one using i
    
        names = driver.find_elements(By.XPATH, '//a[@class="item-title"]') #pull item name elements from website into list
        for n in range(len(names)):
            name_list.append(names[n].text) #convert to text

        materialList = pd.DataFrame(name_list, columns = ['Item Name']) #create DF based on the name_list array
        
        prices = driver.find_elements(By.XPATH, '//li[@class="price-current "]//strong') #pull price element from website into list
        for p in range(len(prices)):
            price_list.append(prices[p].text) #convert to text

        items = driver.find_elements(By.XPATH, '//div[@class="item-cell"]') #pull whole element from website for cross reference
        for i in range(len(items)):
            item_list.append(items[i].text) #convert to text
    
    
    iterator = 0 #separate index iterator for item list, need to check if an item has no listed price or if it does have a price
    for item in range(len(name_list)): 
        if len(price_list) > iterator: #Implement if statement if last items on webpage don't have prices listed
            if (name_list[item] in item_list[item]) and ('$'+price_list[iterator] in item_list[item]): #check if the name and price match the whole item
                price_list[iterator] = price_list[iterator].replace(',', '') #get rid of any , for 1000+ marked items
                materialList.loc[materialList.index[item], 'MSRP Price'] = float(price_list[iterator]) #add price to the index
                iterator += 1 #iterate the item_list index up one
            
            
            #if conditions aren't met, the price column is marked as NaN for the index
    
    return materialList
```

```python
#RUNNING WHILE LOOP FOR INPUT and DF creation

#polls for user input until input is correct
while True:
    userin = UserIn()
    if userin[0] == 'N': #bail condition
        LeaveScreen()
        break
    elif userin[0] == 'Y': #continue
        
        #Create DFs from parts list not bound by platform
        GPUDF = ReadLink(GPU)
        PSUDF = ReadLink(PSU)
        MEMDF = ReadLink(MEM)
        HDDDF = ReadLink(HDD)
        SSDDF = ReadLink(SSD)
        
        budget = PickAPart('Enter your ideal budget:') #Choose budget
        
        while True:
            platform = PickAPart('Pick a Platform: AMD or Intel?:') #Choose a platform
            if platform[0] == 'AMD' or platform[0] == 'Intel': #make sure input is correct
                break
            else:
                ErrorCall() #Calls if input isn't correct
        
        while True:
            coolType = PickAPart('Pick a CPU Cooling Method: Air or Water?:') #Choose a cooling type
            if coolType[0] == 'Air' or coolType[0] == 'Water': #make sure input is correct
                break
            else:
                ErrorCall() #Calls if input isn't correct
        
        if platform[0] == 'AMD': #Create DFs of every part type under AMD
            CPUDF = ReadLink(CPU_AMD)
            MOBDF = ReadLink(MOB_AMD)
            
            if coolType[0] == 'Air': #Create DF for AMD Air type
                COLDF = ReadLink(AIR_AMD)
            else: #Create DF For AMD Water Type
                COLDF = ReadLink(WTR_AMD)
                
        else: #Create DFs of every part type under Intel
            CPUDF = ReadLink(CPU_INT)
            MOBDF = ReadLink(MOB_INT)
            
            if coolType[0] == 'Air': #Create DF for Intel Air type
                COLDF = ReadLink(AIR_INT)
            else: #Create DF For Intel Water Type
                COLDF = ReadLink(WTR_INT)
            
        break
    else:
        ErrorCall() #default in case initial input is incorrect
```

```python
GUI(CPUDF, MOBDF, GPUDF, PSUDF, MEMDF, HDDDF, SSDDF, COLDF, budget[0]) #run GUI code
```
